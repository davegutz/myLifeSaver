```python
import pandas as pd
from sklearn.linear_model import LinearRegression  #scikit-learn
import numpy as np
import statsmodels.api as sm

# Load the CSV file
# Assuming a file 'data.csv' with columns: 'input1', 'input2', 'output'
df = pd.read_csv('lhs_taylor_results.csv')
print("Data Preview:")
print(df.head())

# Define Input (X) and Output (y) Variables
# Map multiple inputs to one output
X = df[['yrs_il_single', 'yrs_il_double', 'yrs_sum_al']] 
y = df['added_lc_worth_norm']

# Add a constant to the independent variables (for intercept)
X = sm.add_constant(X)

# Initialize and Train the Model
# model = LinearRegression()
# model.fit(X, y)

# # 6. Display Results
# print("\nRegression Results:")
# print(f"Intercept (beta0): {model.intercept_}")
# print(f"Coefficients (beta1, beta2): {model.coef_}")

# Perform the multivariable regression
model = sm.OLS(y, X).fit()

# Display results, including p-factors (p-values)
print("\nRegression Summary:")
print(model.summary())

# Extract p-values for specific evaluation
p_values = model.pvalues
print("\nP-values for each coefficient:")
print(p_values)

# Interpret accuracy
print("\nInterpretation:")
for feature, p_val in p_values.items():
    if p_val < 0.05:
        print(f" - {feature}: Significant (p={p_val:.4f})")
    else:
        print(f" - {feature}: Not significant (p={p_val:.4f})")

```
